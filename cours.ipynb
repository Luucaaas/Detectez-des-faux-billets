```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ozone = pd.read_csv("Data_cours/Dataset_ozone.csv",delimiter=';')
```

```python
ozone
```

```python
plt.scatter(ozone["T12"], ozone["maxO3"], color='blue', marker='o', label='Data Points')

plt.xlabel('T12')
plt.ylabel('maxO3')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(ozone['T12'], ozone['maxO3'], color='black')

# Ajouter des Ã©tiquettes et un titre
plt.xlabel('T12', fontsize=14)
plt.ylabel('maxO3', fontsize=14)
plt.title('Scatter plot de T12 vs maxO3', fontsize=16)

# Ajouter une grille pour amÃ©liorer la lisibilitÃ©
plt.grid(True)

# Afficher le graphique
plt.show()
```

```python
sns.set()

ax = sns.scatterplot(x="T12", y="maxO3", data=ozone)
ax.set(xlabel='T12', ylabel='MaxO3')
```
